```python
import pandas as pd
import os
```

```python
diretorio_atual = os.getcwd()
```

```python
df = pd.read_csv('fonte_dados.csv', delimiter=';', encoding='utf-8')
df['DataContratacao'] = pd.to_datetime(df['DataContratacao'], format='%Y-%m-%d')
df['Salario'] = df['Salario'].astype(str)
df
```

```python
df.dtypes
```

```python
df[df.duplicated(keep=False)]
```

```python
df = df.drop_duplicates(keep='first')
```

```python
df_masculino = df[df['Sexo'] == 'Masculino']
df_feminino = df[df['Sexo'] == 'Feminino']
df_escriturario = df[df['Cargo'] == 'Escriturário']
df_assessor = df[df['Cargo'].str.startswith('Assessor')]
df_gerentes = df[df['Cargo'].str.startswith('Gerente ')]
```

```python
dataframes = [df_masculino, df_feminino, df_escriturario, df_assessor, df_gerentes]

for nome_df, dataframe in zip(['Funcionarios_Masculinos.csv', 'Funcionarios_Femininos.csv', 'Funcionarios_Escriturarios.csv', 'Funcionarios_Assessores.csv', 'Funcionarios_Gerentes.csv'], dataframes):

    caminho_arquivo_csv = os.path.join(diretorio_atual, nome_df)
    
    dataframe.to_csv(caminho_arquivo_csv, index=False)
```
